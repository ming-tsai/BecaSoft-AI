Para los datos de más abajo, aplicar Cluster Analysis para determinar cuáles son los puntos que forman los 4 grupos. Construya un script en Python para producir los siguientes gráficos:


```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Lee datos y los grafica
df = pd.read_csv("exercise-1-data.csv")
plt.scatter(df['z'],df['y'], c='b', s=40)
plt.xlabel("Z")
plt.ylabel("Y")

#---create a matrix containing all points---
X = np.array(list(zip(df['z'],df['y'])))
```

```python
#---using sci-kit-learn---
from sklearn.cluster import KMeans
k = 4
kmeans = KMeans(n_clusters=k)

kmeans = kmeans.fit(X)

labels = kmeans.predict(X)

centroids = kmeans.cluster_centers_

print(labels)
print(centroids)

#---map the labels to colors---
c = ['b','r','y','g','c','m']
colors = [c[i] for i in labels]

plt.scatter(df['z'],df['y'], c=colors, s=18)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=100, c='black')
plt.xlabel("Z")
plt.ylabel("Y")
```
